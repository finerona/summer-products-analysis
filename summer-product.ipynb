```python
newdf = df[['shipping_is_express', 'countries_shipped_to', 'price', 'units_sold', 'origin_country','theme','uses_ad_boosts']] 
## selecting the different columns needed for data analysis
```

```python
newdf.head()
```

```python
newdf['sales'] = newdf['price'] * newdf['units_sold'] ## creating  a new column sales 
```

```python
newdf.head()
```

```python
data={'0':'NO','1':'YES'} #mapping, this is to replace 0 and 1 with YES and NO
```

```python
newdf['uses_ad_boosts']=newdf['uses_ad_boosts'].astype(str).map(data)
newdf['shipping_is_express']=newdf['shipping_is_express'].astype(str).map(data)
```

```python
newdf.head()
```

```python
newdf.dtypes #checking the data types
```

```python
 newdf.isna().sum() ##checking for nan values 
```

```python
newdf['origin_country'] = newdf['origin_country'].astype(str).replace('nan', 'Not specified') # replacing the nan values with non specified
```

```python
newdf.to_csv('Cleanedsumddata.csv', index=False) # saving the data to csv
```
